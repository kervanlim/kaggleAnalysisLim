## **Pre-processing**

The first things I want to do is drop feature Arrival Delay in Minutes because it's redundant

```python
df_pre = df.drop(columns=['Arrival Delay in Minutes']).copy()
df_pre.columns
```

```python
df.head()
```

In this section, I'm trying to do Feature Extraction by changing the value of every categorical Feature

```python
df_pre['satisfaction'] = df_pre['satisfaction'].astype('category').cat.codes
df_pre['Gender'] = df_pre['Gender'].astype('category').cat.codes
df_pre['Customer Type'] = df_pre['Customer Type'].astype('category').cat.codes
df_pre['Type of Travel'] = df_pre['Type of Travel'].astype('category').cat.codes
dummy_Class = pd.get_dummies(data=df_pre['Class'], prefix='is')
df_pre = df_pre.join(dummy_Class)
df_pre.head()
```

```python
df_pre = df_pre.drop(columns=['Class'])
df_pre.head()
```

Drop the outliers using Z-Score because the data distribution is close to normal and IQR for abnormal.

```python
import scipy.stats as stats

print('Before: ', len(df_pre))
filtered_entries = np.array([True]* len(df_pre))

z_score = abs(stats.zscore(df_pre['Age']))
filtered_entries = (z_score < 3) & filtered_entries

df_pre = df_pre[filtered_entries]
print('After: ', len(df_pre))
```

```python
print('Before: ', len(df_pre))
filtered_entries_IQR = np.array([True]* len(df_pre))

for i in ['Flight Distance', 'Departure Delay in Minutes']:
    Q1 = df_pre[i].quantile(0.25)
    Q3 = df_pre[i].quantile(0.75)
    IQR = Q3-Q1
    low_limit = Q1 - (1.5 * IQR)
    high_limit = Q3 + (1.5* IQR)
    filtered_entries_IQR = ((df_pre[i] >= low_limit) & (df_pre[i] <= high_limit)) & filtered_entries_IQR
    
df_pre = df_pre[filtered_entries_IQR]
print('After: ', len(df_pre))
```

Normalization/Standardization

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df_pre['Age_Std'] = StandardScaler().fit_transform(df_pre['Age'].values.reshape(len(df_pre),1))
df_pre['Flight_Dist_Norm'] = MinMaxScaler().fit_transform(df_pre['Flight Distance'].values.reshape(len(df_pre),1))
df_pre['Dprtr_Delay_Norm'] = MinMaxScaler().fit_transform(df_pre['Departure Delay in Minutes'].values.reshape(len(df_pre),1))
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(df_pre.corr(), cmap='Blues', annot=True, fmt='.2f')
```

After finishing preprocessing step, I use heatmap to select the features to be used in the model. The features are **Customer Type, Seat Comfort, Inflight entertainment, Online support, Ease of Online Booking, On-board service, Leg room service, Baggage handling, Checkin service, Cleanliness, Online boarding, is_Business, is_Eco, is_Eco Plus.**

```python
df_pre[['satisfaction']].value_counts()
```

```python
from imblearn import over_sampling

X = df_pre.drop(columns=['satisfaction','Gender','Type of Travel','Departure/Arrival time convenient','Food and drink','Gate location','Age_Std','Flight_Dist_Norm', 'Dprtr_Delay_Norm', 'Flight Distance','Age','Departure Delay in Minutes'])
y = df_pre['satisfaction'].values

X_SMOTE, y_SMOTE = over_sampling.SMOTE().fit_resample(X,y)
pd.Series(y_SMOTE).value_counts()
```

## **Train Model**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

def eval_model(X_train, y_train, X_test, y_test, y_pred, model):
    print('Accuracy: ', accuracy_score(y_test, y_pred))
    print('Precision: ', precision_score(y_test, y_pred))
    print('Recall: ', recall_score(y_test, y_pred))
    print('Train Score: ', model.score(X_train, y_train))
    print('Test Score: ', model.score(X_test, y_test))
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X_SMOTE, y_SMOTE, test_size=0.3, random_state=42)
print('Train: ', len(y_train))
print('Test: ', len(y_test))
```

### **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

y_pred_lr = lr_model.predict(X_test)
eval_model(X_train, y_train, X_test, y_test, y_pred_lr, lr_model)
```

Underfit model, try tuning hyperparameter

```python
penalty = ['l1','l2','elasticnet']
C = [0.0001, 0.001, 0.002]
hyperparameters_lr = dict(penalty=penalty, C=C)

lr_model_tuning = LogisticRegression()
best_model_lr = RandomizedSearchCV(lr_model_tuning, hyperparameters_lr, cv=5, random_state=42, scoring='precision')
best_model_lr.fit(X_train, y_train)

y_pred_lr = best_model_lr.predict(X_test)
eval_model(X_train, y_train, X_test, y_test, y_pred_lr, best_model_lr)
```

Still underfit, try another model

### **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier

KNN_Model = KNeighborsClassifier()
KNN_Model.fit(X_train, y_train)

y_pred_KNN = KNN_Model.predict(X_test)
eval_model(X_train, y_train, X_test, y_test, y_pred_KNN, KNN_Model)
```

Not too overfit, but try tuning the hyperparameter

```python
n_neighbors = list(range(1,30))
algorithm = ['auto', 'ball_tree', 'kd_tree', 'brute']
p = [1,2]
hyperparameters = dict(n_neighbors=n_neighbors, algorithm=algorithm, p=p)

KNN_Model_tuning = KNeighborsClassifier()
best_model_KNN = RandomizedSearchCV(KNN_Model_tuning, hyperparameters, cv = 5, random_state=42, scoring='precision')
best_model_KNN.fit(X_train, y_train)

y_pred_KNN_tuning = best_model_KNN.predict(X_test)
eval_model(X_train, y_train, X_test, y_test, y_pred_KNN_tuning, best_model_KNN)
```

If we look at the precision and the range between train and test score, it is already a good model, but KNN is really bad if the data too big. According to this article (https://levelup.gitconnected.com/knn-failure-cases-limitations-and-strategy-to-pick-right-k-45de1b986428), KNN is a distance-based algorithm, the cost of calculating distance between a new point and each existing point is very high which in turn degrades the performance of the algorithm. I will consider to use KNN algorithm if others algorithm get a bad evaluation score.

## **Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

y_pred_dt = dt_model.predict(X_test)
eval_model(X_train, y_train, X_test, y_test, y_pred_dt, dt_model)
```

A little bit overfit, try tuning hyperparameters

```python
criterion = ['gini','entropy']
splitter = ['best','random']
max_depth = [int (x) for x in np.linspace(1,100, num=30)]
min_samples_split = [2,5,10,100]
min_samples_leaf = [1,2,4,10,20,50]
hyperparameters_dt = dict(criterion=criterion,splitter=splitter, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf)

dt_model_tuning = DecisionTreeClassifier()
best_dt_model = RandomizedSearchCV(dt_model_tuning, hyperparameters_dt, cv=5, random_state=42, scoring='precision')
best_dt_model.fit(X_train, y_train)

y_pred_dt_tuning = best_dt_model.predict(X_test)
eval_model(X_train, y_train, X_test, y_test, y_pred_dt_tuning, best_dt_model)
```

It's pretty good score. We will try last algorithm, RandomForest

### **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(criterion='entropy',min_samples_split=10,max_depth=59, min_samples_leaf=1, random_state=42)
rf.fit(X_train,y_train)

y_pred_rf = rf.predict(X_test)
eval_model(X_train, y_train, X_test, y_test, y_pred_rf, rf)
```

```python
from sklearn.ensemble import RandomForestClassifier

criterion = ['gini','entropy']
max_depth = [int (x) for x in np.linspace(1,100, num=30)]
min_samples_split = [2,5,10,100]
min_samples_leaf = [1,2,4,10,20,50]
hyperparameters_rf = dict(criterion=criterion,max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf)

rf_tuning = RandomForestClassifier()
best_model_rf = RandomizedSearchCV(rf_tuning, hyperparameters_rf, cv=5, random_state=42, scoring='precision')
best_model_rf.fit(X_train,y_train)

y_pred_rf_tuning = best_model_rf.predict(X_test)
eval_model(X_train, y_train, X_test, y_test, y_pred_rf_tuning, best_model_rf)
```

```python
print('Best Criterion: ', best_model_rf.best_estimator_.criterion)
print('Best max_depth: ', best_model_rf.best_estimator_.max_depth)
print('Best min_samples_split: ', best_model_rf.best_estimator_.min_samples_split)
print('Best min_samples_leaf: ', best_model_rf.best_estimator_.min_samples_leaf)
```
